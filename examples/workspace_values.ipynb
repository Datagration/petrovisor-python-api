```python
import petrovisor as pv
```

# Authorization

```python
# url
# discovery_url = r'https://identity.eu1.petrovisor.com'
discovery_url = r'https://identity.us1.petrovisor.com'

# workspace
workspace = 'Workspace Name'

# generate access key (only for the first time, comment the line afterwards)
key = pv.PetroVisor.generate_credentials_key()
# print(f"key : '{key}'")
# save key for later to reuse
# key = ''

api = pv.PetroVisor(workspace = workspace, discovery_url = discovery_url, key = key)
print(api.Api)
```

# Examples: Workspace Values

### 'Numeric', 'NumericWithUnit', 'String', 'Dictionary', 'List', 'Enumeration'

```python
# get workspace value names
api.get_workspace_value_names()
```

```python
# get workspace values
api.get_workspace_values()
```

```python
# get workspace values of specified type
# possible type: 'Numeric', 'NumericWithUnit', 'String', 'Dictionary', 'List', 'Enumeration'
api.get_workspace_values(value_type='Dictionary')
```

```python
# add Numeric value
api.add_workspace_value('Python Numeric', 1)
```

```python
# add Numeric value with unit
api.add_workspace_value('Python NumericWithUnit', 1, unit = 'm')
```

```python
# add String value
api.add_workspace_value('Python String', 'Hello PetroVisor!')
```

```python
# add List value
api.add_workspace_value('Python List', [1,'2','three',])
```

```python
# add Dictionary value
api.add_workspace_value('Python Dictionary', {'a': 1, 'b': [1,2,3], 'c': 'foo'})
```

```python
# add Enumeration value
api.add_workspace_value('Python Enumeration', tuple({'a': 1, 'b': 2}.items()))
```

```python
# get workspace value
api.get_workspace_value('Python Dictionary')
```

```python
# delete workspace value
api.delete_workspace_value('Python Dictionary')
```
